```python
data_technical=data_master.copy()
```

```python
s = data_technical['Close']

def Average_SMA(data,prices,window):
    for i in window:
        data['sma_'+str(i)]=prices.rolling(window=i).mean()
    return data

def Average_EMA(data,prices,window):
    for i in window:
        data['ema_'+str(i)]=prices.ewm(span=i).mean()
    return data

def RSI(data,prices,window):
    for i in window:
        data['rsi_' + str(i)] = ta.RSI(prices, timeperiod=i)
    return data

def bollingerBands(data,prices,window):
    data_sma=Average_SMA(data=data['Close'],prices=s,window=windows)   
    for i in window:    
        for j in deviation:
            # upper_sigma
            data['upper_' + str(i) + '_dev_' + str(j)] = data_sma['sma_' + str(i)] + s.rolling(window=i).std(ddof=0) * j
            #lower_sigma
            data['lower_' + str(i) + '_dev_' + str(j)] = data_sma['sma_' + str(i)] - s.rolling(window=i).std(ddof=0) * j

            # ボリンジャーバンドの上か下か
            data['upper_check' + str(i) + '_dev_' + str(j)] = s >= data['upper_' + str(i) + '_dev_' + str(j)]
            data['lower_check' + str(i) + '_dev_' + str(j)] = s <= data['lower_' + str(i) + '_dev_' + str(j)]
    return data
```

```python
data_technical=Average_SMA(data=data_technical,prices=s,window=windows)
data_technical=Average_EMA(data=data_technical,prices=s,window=windows)
data_technical=RSI(data=data_technical,prices=s,window=windows)
data_technical=bollingerBands(data=data_technical,prices=s,window=windows)
```

```python
def TrueFalseConversion(data):
    data.replace({False: 0, True: 1}, inplace=True)
    return data

def Data_Extraction(data,start):
    start_timestamp= pd.to_datetime(start)
    data = data.set_index('Date')
    data = data.loc[start_timestamp:, :]
    data.reset_index(drop=False, inplace=True)        
    return data
```

```python
data_technical=TrueFalseConversion(data_technical)
data_output=Data_Extraction(data=data_technical,start='2021-01-01')
data_output.to_csv('data_technical_tentative.csv')
```
