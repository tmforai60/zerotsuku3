```python
import pandas as pd
from pandas_datareader import data
%matplotlib inline
```

```python
start = '2021-01-01'
end = '2022-01-01'
 
hist_prices = data.DataReader('1321.T', 'yahoo', start, end)
```

```python
hist_prices
```

```python
hist_prices = hist_prices[['High','Low','Open','Close']]
```

```python
hist_prices_cal = hist_prices
```

```python
hist_prices_cal
```

```python
ma_day = [10,20,50]

for ma in ma_day:
    column_name = "MA {}".format(str(ma))
    hist_prices[column_name]=hist_prices['Close'].rolling(window=ma).mean()
    
hist_prices[['Close','MA 10','MA 20','MA 50']].plot(subplots=False,figsize=(14,8))
```
