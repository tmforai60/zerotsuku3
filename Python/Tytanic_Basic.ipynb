```python
import pandas as pd

import os
working_directory=os.getcwd()
print(working_directory)
```

```python
import warnings
warnings.simplefilter('ignore')
```

```python
path_train=working_directory+'/train.csv'
path_test=working_directory+'/test.csv'
path_submit=working_directory+'/submit.csv'
df_train=pd.read_csv(path_train)
df_test=pd.read_csv(path_test)
df_submit=pd.read_csv(path_submit)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train['survived'].value_counts()
```

```python
df_train=df_train[["survived","sibsp","parch","fare"]]
```

```python
df_train
```

```python
y=df_train["survived"] #目的変数
x=df_train.drop("survived",axis=1) #目的変数の除いたデータ
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x,y)
```

```python
#説明変数をdf_trainと揃える
df_test=df_test[["sibsp","parch","fare"]]
```

```python
df_test
```

```python
pred=model.predict_proba(df_test)[:,1]
```

```python
print(pred[:5])
```

```python
#予測結果の投稿
```

```python
df_submit['y']=pred
```
